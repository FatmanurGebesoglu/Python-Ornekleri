```python
import pandas as pd
import numpy as np
```

```python
sozlukVerisi={"istanbul":[30,29,np.nan],"ankara":[20,np.nan,25],"izmir":[40,39,28]}
havaDurumuDataFrame=pd.DataFrame(sozlukVerisi)
```

```python
havaDurumuDataFrame
```

```python
havaDurumuDataFrame.dropna()
#satırda eksikse varsa siler
```

```python
havaDurumuDataFrame.dropna(axis=1)
#sutunda eksik varsa siler
```

```python
yeniVeri={"istanbul":[30,29,np.nan],"ankara":[20,np.nan,25],"izmir":[40,39,28],"antalya":[45,np.nan,np.nan]}
yeniVeriDataFrame=pd.DataFrame(yeniVeri)
```

```python
yeniVeriDataFrame
```

```python
yeniVeriDataFrame.dropna(axis=1,thresh=2)
# thresh=2 kodu 2 ve üzeri nan var ise drop et anlamına gelir
```

```python
yeniVeriDataFrame.fillna(20)
#nan olan yerleri 20 ile doldur anlamına gelir
```

groupby

```python
maasdict= {"departman":["yazilim","yazilim","pazarlama","pazarlama","hukuk","hukuk"],
           "calisan ismi":["ahmet","mehmet","atil","burak","zeynep","fatma"],
           "maas":[100,150,200,300,400,500]}
```

```python
maasDataFrame=pd.DataFrame(maasdict)
```

```python
maasDataFrame
```

```python
grupobjesi=maasDataFrame.groupby("departman")
```

```python
grupobjesi.count()
```

```python
grupobjesi.mean()
```

```python
grupobjesi.describe()
```
